```python
#import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#load Airbnb data
listings = "Resources/listings.csv"
```

```python
#encoding error resolution
listings_df = pd.read_csv(listings, encoding="ISO-8859-1")
```

```python
#dropping rows with no data
listings_df.dropna(how = "all")
```

```python
#review location and cleanliness distribution
plt.scatter(listings_df["review_scores_cleanliness"], listings_df["review_scores_location"], color ="teal",
          edgecolors = "k")

plt.title(f"Location vs. Cleanliness")
plt.xlabel("Cleanliness")
plt.ylabel("Location")
plt.grid(False)
plt.savefig ("images/Location vs. Cleanliness.png", bbox_inches='tight')
```

```python
#looking into the distribution of reviews against number of listings per host
plt.scatter(listings_df["host_total_listings_count"], listings_df["number_of_reviews"], color ="teal",
          edgecolors = "k")

plt.title(f"Total Listings vs. Number of Reviews")
plt.xlabel("Total Listings")
plt.ylabel("Number of Reviews")
plt.grid(False)
plt.savefig ("images/Total Listings vs. Number of Reviews.png", bbox_inches='tight')
```

```python
plt.bar(listings_df["property_type"], listings_df["number_of_reviews"])
plt.xticks(rotation=90)
plt.title("Reviews by Property Type")
plt.xlabel("Property Type")
plt.ylabel("Reviews")
plt.savefig("images/Property Type Review.png", bbox_inches='tight')
```

```python
plt.figure(figsize=(50,10))
plt.bar(listings_df["neighbourhood"], listings_df["number_of_reviews"])
plt.xticks(rotation=90)
plt.title("Reviews by Neighbourhood")
plt.xlabel("Neighbourhood")
plt.ylabel("Reviews")
#plt.tight_layout()
plt.savefig("images/Neighbourhood Review.png", bbox_inches='tight')
```

```python
plt.bar(listings_df["room_type"], listings_df["number_of_reviews"])
plt.title("Reviews by Room Type")
plt.xlabel("Room Type")
plt.ylabel("Reviews")
plt.savefig("images/Room Type Review.png", bbox_inches='tight')
```

```python
plt.bar(listings_df["bed_type"], listings_df["number_of_reviews"])
plt.title("Reviews by Bed Type")
plt.xlabel("Bed Type")
plt.ylabel("Reviews")
plt.savefig("images/Bed Type Review.png", bbox_inches='tight')
```

```python
plt.bar(listings_df["cancellation_policy"], listings_df["number_of_reviews"])
plt.xticks(rotation=90)
plt.title("Cancellation Policy Impact on Host Ratings")
plt.xlabel("Cancellation Policy")
plt.ylabel("Host Rating")
plt.savefig("images/Cancellation Policy.png", bbox_inches='tight')
```
